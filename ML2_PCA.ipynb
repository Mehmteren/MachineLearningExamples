```python
from sklearn import model_selection
```

```python
cv_10 = model_selection.KFold(n_splits=10, shuffle=True, random_state=1)
```

```python
lm = LinearRegression()
```

```python
RMSE = []
```

```python
for i in np.arange(1,X_reduced_train.shape[1]+1):
  score = np.sqrt(-1*model_selection.cross_val_score(lm, X_reduced_train[:,0:i], y_train, cv=cv_10, scoring="neg_mean_squared_error").mean())
  RMSE.append(score)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(RMSE, '-v')
plt.xlabel("Komponent sayÄ±sÄ±")
plt.ylabel("RMSE")
plt.title("MaaÅ tahmin modeli iÃ§in plt modeli")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
diabets = pd.read_csv("/content/drive/MyDrive/Ornek_calÄ±smalar/diabetes.csv", sep=",")
```

```python
df = diabets.copy()
```

```python
df.head()
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
IQR = q3 - q1
```

```python
alt_sinir = q1 - 1.5*IQR
ust_sinir = q3 + 1.5*IQR
```

```python
aykiri=(df< alt_sinir) | (df> ust_sinir)
```

```python
aykiri.head()
```

```python
df[aykiri]
```

```python
aykiri.sum()
```

```python
df_table = pd.DataFrame(df)
```

```python
type(df_table)
```

```python
t_df = df_table[~((df_table< alt_sinir) | (df_table> ust_sinir)).any(axis =1)]
```
